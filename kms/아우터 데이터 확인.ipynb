```python
# 만족도
#  만족도 열을 ' / '로 분리하여 딕셔너리 형태로 변환하는 함수 정의
def split_satisfaction(text):
    if pd.isna(text) or str(text).strip() == "":
        return {}
    parsed_dict = {}
    parts = str(text).split(' / ')
    for part in parts:
        if ': ' in part:
            key, value = part.split(': ', 1)
            parsed_dict[key.strip()] = value.strip()
    return parsed_dict

satisfaction_expanded = df_reviews['만족도'].apply(split_satisfaction).apply(pd.Series)
df_reviews_analysis = pd.concat([df_reviews, satisfaction_expanded], axis=1)

display(df_reviews_analysis.head())

review_list = ['사이즈','화면 대비 색감', '두께감', '퀄리티', '보온성']

for review in review_list:
    if review in df_reviews_analysis.columns:
        counts = df_reviews_analysis[review].value_counts(dropna=False)
        ratios = df_reviews_analysis[review].value_counts(normalize=True, dropna=False) * 100
        
        result_df = pd.DataFrame({
            '개수': counts,
            '비율(%)': ratios.round(2) # 소수점 둘째 자리까지 반올림
        })
        
        print(f"\n========== [ {review} ] ==========")
        print(result_df)

```

```python
# 만족도에서 Nan인 케이스 비교

cols = ['사이즈', '화면 대비 색감', '두께감', '퀄리티', '보온성']

df_reviews_analysis.groupby(
    [df_reviews_analysis[col].notna().rename(col) for col in cols]
).size().reset_index(name='행 개수')
```
