```python
import pandas as pd
import glob
import os

# 1. 파일이 실제 들어있는 폴더 경로 (이미지상 'data' 폴더인 것으로 보입니다)
# 만약 코드가 data 폴더 밖에서 실행된다면 'data/' 로 설정하세요.
path = 'data/' 

def load_and_union(keyword):
    # 파일명에 keyword(products 또는 reviews)가 포함되고 .csv로 끝나는 모든 파일 찾기
    file_pattern = os.path.join(path, f"*{keyword}*.csv")
    all_files = glob.glob(file_pattern)
    
    # 디버깅용: 찾은 파일 목록 출력
    print(f"[{keyword}] 찾은 파일 개수: {len(all_files)}개")
    if len(all_files) == 0:
        print(f"경고: {file_pattern} 경로에서 파일을 찾지 못했습니다.")
        return pd.DataFrame() # 빈 데이터프레임 반환

    # 데이터 로드 및 통합
    df_list = [pd.read_csv(f) for f in all_files]
    df_combined = pd.concat(df_list, axis=0, ignore_index=True)
    
    # 중복 행 제거 (완전히 똑같은 행 하나만 남기기)
    df_combined = df_combined.drop_duplicates()
    
    # 결측치를 'null' 문자열로 채우기
    df_combined = df_combined.fillna('null')
    
    return df_combined

# 2. 실행
products_all = load_and_union("products")
reviews_all = load_and_union("reviews")

# 결과 확인
if not products_all.empty:
    print("products_all 생성 완료! 행 개수:", len(products_all))
if not reviews_all.empty:
    print("reviews_all 생성 완료! 행 개수:", len(reviews_all))
```

```python
# 1. 통합 제품 데이터 저장
products_all.to_csv('data/products_all_2604.csv', index=False, encoding='utf-8-sig')

# 2. 통합 리뷰 데이터 저장
reviews_all.to_csv('data/reviews_all_2604.csv', index=False, encoding='utf-8-sig')

print("CSV 파일 저장 완료!")
```
