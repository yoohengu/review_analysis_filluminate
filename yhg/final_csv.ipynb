```python
import pandas as pd

# ── 데이터 로드 ───────────────────────────────────────────────────────────
df = pd.read_csv('reviews_step2_dedup.csv', low_memory=False, encoding='utf-8-sig')
print(f"로드 완료: {len(df):,}건")
print(f"컬럼 수: {df.shape[1]}개")

# ── is_valid_for_topic 필터링 ─────────────────────────────────────────────
before = len(df)
df_final = df[df['is_valid_for_topic'] == True].copy().reset_index(drop=True)
after = len(df_final)

print(f"\nis_valid_for_topic 필터링")
print(f"  전: {before:,}건 → 후: {after:,}건 (-{before - after:,}건)")

# ── 확인 ──────────────────────────────────────────────────────────────────
print(f"\n[주요 컬럼 확인]")
print(f"  리뷰내용_clean 결측: {df_final['리뷰내용_clean'].isna().sum():,}건")
print(f"  리뷰내용_norm  결측: {df_final['리뷰내용_norm'].isna().sum():,}건")
print(f"  한글_글자수 통계:\n{df_final['한글_글자수'].describe()}")

# ── 저장 ──────────────────────────────────────────────────────────────────
df_final.to_csv('df_final.csv', index=False, encoding='utf-8-sig')
print(f"\n저장 완료: df_final.csv ({len(df_final):,}건)")
```
